```python
# The code defines a function "feature_extraction" that takes a file path as input and returns the mean Mel-frequency cepstral coefficients (MFCCs) of the audio file at that path.
# The function loads the audio file using the librosa library and extracts the MFCCs using the "mfcc" function from the "librosa.feature" module. The MFCCs are calculated on the audio signal "x" with the sample rate "sample_rate". The mean of the MFCCs is calculated along the time axis (i.e., the second dimension of the array), and the result is returned by the function.
# https://python.hotexamples.com/examples/SAECrawlers.items/UrlItem/-/python-urlitem-class-examples.html
```

```python
def feature_extraction(file_path):
    # load the audio file
    x, sample_rate = librosa.load(file_path, res_type='kaiser_fast')
    # extract features from the audio
    mfcc = np.mean(librosa.feature.mfcc(y=x, sr=sample_rate, n_mfcc=50).T, axis=0)
    
    return mfcc
```

```python
# The code below  is iterating over all audio files in the directory "C:/Users/it-it/Desktop/Final BootCamp Project/test actor 1,2/" and extracting features from each audio file using the function feature_extraction(file_path). The extracted features are stored in a dictionary features, with the keys being the audio file paths and the values being the extracted feature values.
# https://stackoverflow.com/questions/64801256/how-to-check-if-a-directory-has-music-files-in-python
# https://towardsdatascience.com/extract-features-of-music-75a3f9bc265d
# https://scikit-learn.org/stable/modules/feature_extraction.html
```

```python
features = {}
directory = 'C:/Users/it-it/Desktop/Final BootCamp Project/test actor 1,2/'
for audio in os.listdir(directory):
    audio_path = directory+audio
    features[audio_path] = feature_extraction(audio_path)
```

```python
# To perform a Fourier transform on the audio signal in python, you can use the numpy library. Here's an example code that calculates the Fourier transform of the audio signal and extracts the frequencies from the resulting complex coefficients:
```

```python
# https://towardsdatascience.com/understanding-audio-data-fourier-transform-fft-spectrogram-and-speech-recognition-a4072d228520#:~:text=Fourier%20Transform%20(FT),-An%20audio%20signal&text=Fourier%20Transform%20is%20a%20mathematical%20concept%20that%20can%20decompose%20a,frequency%20present%20in%20the%20signal.
# https://numpy.org/doc/stable/reference/generated/numpy.fft.fftfreq.html
# https://numpy.org/doc/stable/reference/generated/numpy.absolute.html
```

```python
import numpy as np

def audio_fourier_transform(audio_signal, sample_rate):
    audio_signal_fft = np.fft.fft(audio_signal)
    audio_frequencies = np.fft.fftfreq(len(audio_signal), 1/sample_rate)
    audio_frequencies = np.abs(audio_frequencies)

    return audio_frequencies, audio_signal_fft

# Example usage
audio_signal, sample_rate = librosa.load("C:/Users/it-it/Desktop/Final BootCamp Project/test actor 1,2/03-02-06-02-02-02-06.wav", sr=None)
audio_frequencies, audio_signal_fft = audio_fourier_transform(audio_signal, sample_rate)
```

```python
# In the code above, the audio_fourier_transform function calculates the Fourier transform of the audio signal using the numpy.fft.fft function, and then calculates the frequencies of the complex coefficients using the numpy.fft.fftfreq function. The resulting audio frequencies are the absolute values of the frequencies.
```

```python
# In audio processing, the frequencies are typically represented as a spectrogram, which is a graph showing the energy of each frequency present in a signal over time. To convert these resulting audio frequencies and absolute values of frequencies into Hz frequency, you need to divide each bin number by the total number of bins, and then multiply by the sample rate of the signal.
# For example, if the spectrogram has N bins and was computed from a signal with a sample rate of Fs Hz, the frequency corresponding to each bin can be calculated as:
# f = (bin number * Fs) / N
# where f is the frequency in Hz. The absolute value of the frequency corresponds to the magnitude of the signal at that frequency, which can be used to determine the volume of the corresponding tone in the signal.
```

```python
# # The frequency in Hz can be calculated as follows:
# The np.fft.fftfreq function returns the frequencies corresponding to each bin of the FFT, but they are not yet in Hz. To convert them to Hz, you need to multiply them by the sample rate. Note that np.abs is used to obtain the absolute value of the frequencies, because the output of np.fft.fftfreq may be complex.
```

```python
N = len(audio_signal)
audio_frequencies = np.fft.fftfreq(N, 1/sample_rate)

# Convert the frequencies to Hz
audio_frequencies = np.abs(audio_frequencies) * sample_rate
```

```python
import os
import librosa
import pandas as pd

folder_path = "C:/Users/it-it/Desktop/Final BootCamp Project/test actor 1,2/"

mean_frequencies = []

for filename in os.listdir(folder_path):
    # Load the audio file
    file_path = os.path.join(folder_path, filename)
    audio, sr = librosa.load(file_path)

    # Calculate the mean frequency
    mean_frequency = librosa.feature.mfcc(audio, sr=sr).mean()
    mean_frequencies.append(mean_frequency)

# Store the results in a dataframe
df = pd.DataFrame({'filename': os.listdir(folder_path), 'mean_frequency': mean_frequencies})

# Print the dataframe
print(df)
```
